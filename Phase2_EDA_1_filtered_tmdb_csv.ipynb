Dropping the unnamed columns

```python
selected_columns = ['genre_ids', 'id', 'original_language',
       'original_title', 'popularity', 'release_date', 'title', 'vote_average',
       'vote_count', 'genre']

df = df[selected_columns].copy()
```

```python
df.head(3)
```

```python
df_copy = df.copy()
```

```python
df_copy['genre'].nunique()
```

```python
df_copy['genre'].unique()
```

```python
df_copy[['main_genre', 'subgenre_1', 'subgenre_2', 'subgenre_3', 'subgenre_4', 'subgenre_5', 'subgenre_6']] = df_copy['genre'].str.split(',', expand=True)
```

```python
df_copy.head(3)
```

```python
# remove punctuation from main_genre column
df_copy['main_genre'] = df_copy['main_genre'].str.replace('[^\w\s]','', regex=True)
```

```python
# sanity check
df_copy.head(3)
```

####Now for the other columns:

```python

df_copy['subgenre_1'] = df_copy['subgenre_1'].str.replace('[^\w\s]','', regex=True)
```

```python
df_copy['subgenre_2'] = df_copy['subgenre_2'].str.replace('[^\w\s]','', regex=True)
```

```python
df_copy['subgenre_3'] = df_copy['subgenre_3'].str.replace('[^\w\s]','', regex=True)
```

```python
df_copy['subgenre_4'] = df_copy['subgenre_4'].str.replace('[^\w\s]','', regex=True)
```

```python
df_copy['subgenre_5'] = df_copy['subgenre_5'].str.replace('[^\w\s]','', regex=True)
```

```python
df_copy['subgenre_6'] = df_copy['subgenre_6'].str.replace('[^\w\s]','', regex=True)
```
